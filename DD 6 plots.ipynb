```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df_encoded = pd.read_csv("DD_Python_2.csv")

# Reconstruct Engagement_with_FH if needed
# (skip this if already done, with 'No engagement' included)

# Define colors for Outcome
outcome_colors = {
    'Attended': '#A9D18E',  # green
    'Booked but did not attend': '#ED7D31',  # orange
    'Declined': '#C00000'  # red
}

# Variables to plot against Outcome
variables = ['Engagement_with_FH', 'Ethnicity', 'Employment_Status',
             'IDACI', 'Children_in_Family', 'Family_Composition']

plt.figure(figsize=(15, 18))

for i, var in enumerate(variables, 1):
    plt.subplot(3, 2, i)
    
    if df_encoded[var].dtype.name == 'category' or df_encoded[var].dtype == object:
        # Categorical: normalized stacked bar
        counts = df_encoded.groupby(var)['Outcome'].value_counts(normalize=True).unstack(fill_value=0)
        counts.plot(kind='bar', stacked=True, color=[outcome_colors.get(x, 'grey') for x in counts.columns], ax=plt.gca())
        plt.ylabel('Proportion')
        plt.xlabel(var)
    else:
        # Numeric: boxplot
        sns.boxplot(
            x=var, 
            y='Outcome', 
            data=df_encoded, 
            hue='Outcome',          # assign hue
            dodge=False,            # avoid splitting boxes
            palette=outcome_colors,
            legend=False,           # avoid duplicate legend
            ax=plt.gca()
        )
        plt.xlabel(var)
        plt.ylabel('Outcome')
        
    plt.title(f'{var} vs Outcome')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df_encoded = pd.read_csv("DD_Python_2.csv")

# Ensure Outcome is categorical
df_encoded['Outcome'] = pd.Categorical(
    df_encoded['Outcome'],
    categories=['Attended', 'Booked but did not attend', 'Declined'],
    ordered=True
)

# Define variables for plots
variables = ['Engagement_with_FH', 'Ethnicity', 'Employment_Status',
             'IDACI', 'Children_in_Family', 'Family_Composition']

# Define color palette
palette_colors = {
    'Attended': '#A9D18E',  # green
    'Booked but did not attend': '#ED7D31',  # orange
    'Declined': '#C00000'  # red
}

# Save plots
for var in variables:
    plt.figure(figsize=(6,4))
    if df_encoded[var].dtype.name == 'category' or df_encoded[var].dtype == object:
        counts = df_encoded.groupby(var)['Outcome'].value_counts(normalize=True).unstack(fill_value=0)
        counts.plot(kind='bar', stacked=True, color=[palette_colors.get(x, 'grey') for x in counts.columns])
        plt.ylabel('Proportion')
        plt.xlabel(var)
    else:
        sns.boxplot(x='Outcome', y=var, data=df_encoded, palette=palette_colors)
        plt.ylabel(var)
        plt.xlabel('Outcome')
    plt.title(f'{var} vs Outcome')
    plt.tight_layout()
    plt.savefig(f'{var}_vs_Outcome.png', dpi=150)
    plt.close()
```
