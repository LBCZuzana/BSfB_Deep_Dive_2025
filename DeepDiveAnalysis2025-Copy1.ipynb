**Deep Dive BSfB Analysis 2025**

# Introduction

This analysis explores key factors influencing whether families attend, book but do not attend, or decline BSfB appointments.  
The dataset contains information on family demographics, factors and engagement levels, with outcomes recorded as:

- **Attended**  - 30 families
- **Booked but did not attend**  - 30 families
- **Declined** - 40 families

The aim of this deep dive is to better understand how different family characteristics and barriers are associated with these outcomes.  
By exploring patterns and trends, we can identify where engagement strategies are working well and where further support may be needed.

The analysis is structured as follows:
1. A brief summary of the multinomial logistic regression results - looking at family demographics and factors
2. A closer look at six key variables, each supported by visualisations and commentary
3. Correlation of barriers and contextual variables
4. A concluding section summarising findings and implications



## 2. Summary of Multinomial Logistic Regression

A multinomial logistic regression was run to model the likelihood of three possible outcomes:
- **Attended** (reference outcome)  
- **Booked but did not attend**  
- **Declined**

The predictors included a range of factors and demographic variables, such as:  
- Engagement with Family Hubs services (session attendance)  
- Child's ethnicity  
- Maternal employment status at birth  
- IDACI (Index of Deprivation Affecting Children)  
- Number of children in the family  
- Family composition
- Distance from home to the local centre in meters 

### Key findings:
- **Engagement with Family Hubs** was one of the strongest predictors. Families with no or low engagement were far more likely to decline or fail to attend compared to those with good engagement.  
- **Deprivation (IDACI)** suggested that families in less deprived areas were more likely to decline.  
- **Family composition** showed that single-parent households were at greater risk of declining or missing booked appointments.
- Families who **attended** generally lived closer to the centre

These results provide statistical evidence that engagement, geographic and social barriers significantly influence outcomes.

## 4. Correlation of barriers and contextual variables
We also examined relationships among a set of contextual/barrier variables:  
- **CSSW involvement** - 8 families
- Other **agencies involved** (for example, Camden Baby Feeding Team, Opthamology, Physiotherapy) - 52 families
- **Parental mental health** (primary caregivers) - 10 families
- **Child’s physical health needs** - 10 families
- **Parent’s physical health needs** - 5 families
- **NICU/NNU experience** - 10 families
- **Maternal first language** - 7 families
- **Other barriers**  (for example, overcrowding, parental conflict, financial stressors) - 25 families

<img src="https://raw.githubusercontent.com/LBCZuzana/BSfB_Deep_Dive_2025/main/Odds_Ratios_vs_Outcome.png" width="600">


  
### 4.1 How the variables relate to each other
We inspected associations among key barriers/context features:
- **CSSW involvement** tended to co-occur with **multiple agencies involved**, reflecting clustering of statutory support.
- **Parental mental health** showed overlap with **other barriers**, indicating broader complexity when mental health needs are present.
- **Child and parent physical health needs** were positively related, suggesting health vulnerabilities can cluster within families.
- **NICU/NNU** and **Maternal first language** behaved more like **standalone** needs (weak correlations with the rest).

We also checked simple correlations with the encoded outcome:
- **Children in family** had the strongest correlation with outcome *(r ≈ 0.166)*.
- **IDACI (deprivation)** had a smaller but positive correlation *(r ≈ 0.091)*, indicating higher deprivation aligns with worse engagement outcomes.

> Interpretation note: these correlations describe **pairwise associations**, not causal effects.

---

### 4.2 Impact on outcomes (Multinomial logistic regression, odds ratios)
The table below summarises how each barrier is associated with the odds of each outcome (holding the others constant).  
**Odds Ratio - OR > 1** → higher odds of that outcome; **OR < 1** → lower odds.

| Variable | Attended OR | Booked but did not attend OR | Declined OR | Takeaway |
|---|---:|---:|---:|---|
| **CSSW** | **1.11** | **1.67** | **0.54** | CSSW involvement relates to being **in the system** (booked/attend) rather than declining. |
| **Agencies involved** | 1.01 | 0.98 | 1.00 | Little independent effect once other factors are controlled. |
| **Parental mental health** | **1.65** | **1.96** | **0.31** | Strong engagement signal: higher odds to attend or at least book; markedly **lower** odds of decline. |
| **Child physical health needs** | **1.45** | **1.21** | **0.57** | Health needs align with **greater service use** and lower decline. |
| **Parent physical health needs** | **2.47** | **0.75** | **0.54** | Strongest boost to **attendance**; lower odds of booked-only and decline. |
| **NICU/NNU** | **0.72** | 1.02 | **1.36** | Slightly **lower** attendance, **higher** decline; limited effect on booked-only. |
| **Maternal first language** | **1.82** | **0.65** | 0.85 | Where a language need is flagged, odds of **attendance** are higher and **booked-only** lower (may reflect targeted support). |
| **Other barriers** | **0.39** | **1.61** | **1.60** | Broad, compounding barriers reduce attendance and raise both booked-only and decline. |

**Practical reading:**
- **Boosters of attendance:** Parent physical health (OR 2.47), maternal first language flagged (1.82), parental mental health (1.65), child health needs (1.45).  
- **Risks for “booked but did not attend”:** CSSW (1.67), parental mental health (1.96), other barriers (1.61).  
- **Risks for decline:** Other barriers (1.60) and NICU/NNU (1.36) increase decline; parental/child health needs and CSSW reduce it.

> Interpretation note: These are **associations** from a multinomial model. They indicate where odds shift, not causality. Some “counter-intuitive” patterns (e.g., higher attendance with flagged needs) likely reflect **prioritisation and support pathways** that enable engagement.
### Key Findings
- **CSSW involvement** strongly correlated with **agencies involved**, reflecting the clustering of statutory service support.  
- **Parental mental health** correlated with **other barriers** and modestly with **CSSW involvement**, underlining the overlap between mental health challenges and wider social difficulties.  
- **Physical health needs (child)** correlated with **parent physical health needs**, suggesting clustering of health vulnerabilities within families.  
- **NICU/NNU experience** showed little overlap with other variables — more of an isolated barrier.  
- **Maternal first language** showed weak correlations with other barriers, meaning language needs are relatively distinct.  

### Implications
- Families involved with statutory or multiple agencies often experience **clusters of overlapping barriers**.  
- Some barriers (e.g., language, NICU experience) are more **standalone** but still require targeted support.  
- Interventions should account for both types: **multi-barrier families** needing holistic support, and **single-barrier families** needing tailored solutions.

## 5. Conclusion

This deep dive analysis highlights several important themes:

1. **Engagement with Family Help** is the single strongest driver of positive outcomes.  
2. **Socioeconomic disadvantage** (deprivation, unemployment, lone parenting, large families) consistently reduces attendance.  
3. **Barriers cluster** in families with statutory service involvement, particularly around mental health and social support needs.  
4. **Isolated barriers** (like language or NICU experience) still present risks but require different strategies than clustered barriers.  
5. For families with known barriers, following a booking, **an ongoing engagement and tailored support are to be considered**.  

### Overall Insights
- Outcomes are shaped by the **interaction between engagement and structural disadvantage**.  
- Families with multiple pressures (deprivation, unemployment, single parenting, many children) are at highest risk of disengagement.  
- Culturally and logistically tailored strategies can help address both clustered and standalone barriers.  

### Next Steps
- Build a **predictive model** that combines these variables to flag families at risk of non-attendance or decline.  
- Develop **targeted engagement strategies** for deprived areas, larger families, families living further away from the centre, and single-parent households.  
- Gather **qualitative insights** from families to understand lived experiences behind the patterns observed.
